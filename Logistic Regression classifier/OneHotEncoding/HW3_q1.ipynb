Question 1: Predicting Heart Disease

```python
# a) reading the data from the link given
import pandas as pd

df = pd.DataFrame()

df=pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv")
# b)
df[0::20]
```

```python
#c) creating the feature columns

feature_cols=['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']

# use the above list to select the features from the original DataFrame
X = df[feature_cols]  
# label vector is the last column of AHD
y = df['AHD']
y[::20]
```

 # D)


```python


from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test =train_test_split(X,y,test_size=0.25,random_state=6)
```

```python
#KNN
# part e)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# In the following line, "knn" is instantiated as an "object" of KNeighborsClassifier "class". 

k = 3
knn = KNeighborsClassifier(n_neighbors=k) 
# We can use the method "fit" of the "object knn" along with training dataset and labels to train the model.

knn.fit(X_train, y_train)
y_predict_kn = knn.predict(X_test)
score_kn = accuracy_score(y_test,y_predict_kn)
print(score_kn)
```

```python
# "my_logreg" is instantiated as an "object" of LogisticRegression "class". 
# "my_decisiontree" is instantiated as an "object" of DecisionTreeClassifier "class". 
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

my_logreg = LogisticRegression(random_state=5)

my_decisiontree = DecisionTreeClassifier(random_state=5)
```

```python
# We can use the method "fit" of the objects "my_logreg" and "my_decisiontree" along with training dataset and labels to train the model.

my_logreg.fit(X_train, y_train)

my_decisiontree.fit(X_train, y_train)
```

```python
# testing


from sklearn.metrics import accuracy_score

y_predict_lr = my_logreg.predict(X_test)  # lr is abbreviation for logistic regreesion
y_predict_dt = my_decisiontree.predict(X_test)  # dt is abbreviation for decision tree


score_lr = accuracy_score(y_test, y_predict_lr)
score_dt = accuracy_score(y_test, y_predict_dt)


print(score_lr)
print(score_dt)   # printing accuracy of all three classes which are logistic,decision tree and knn
```

#  F)


```python
# categorical features
# Gender m=0, f=1
# ChestPain: typical=0, asymptomatic=1, nonanginal=2, nontypical=3
# Thal: fixed=0, normal=1, reversable=2

data_Frame = pd.get_dummies(df, columns = ["Gender", "ChestPain", "Thal"])


#printing first few rows
data_Frame.head()
```

# G)

```python
# we have to repeat the process for D) and E)
# doing D)
feature_updated=['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak','Gender_f','Gender_m','ChestPain_asymptomatic',
                'ChestPain_nonanginal','ChestPain_nontypical', 'ChestPain_typical','Thal_fixed','Thal_normal','Thal_reversable']

# use the above list to select the features from the original DataFrame
Xnew = data_Frame[feature_updated]  
# label vector is the last column of AHD
ynew = data_Frame['AHD']

Xnew_train, Xnew_test, ynew_train,ynew_test =train_test_split(Xnew,ynew,test_size=0.25,random_state=6)
```

```python
# doing E)
# knn
k = 3
knn = KNeighborsClassifier(n_neighbors=k) 
# We can use the method "fit" of the "object knn" along with training dataset and labels to train the model.

knn.fit(Xnew_train, ynew_train)
y_predict_kn = knn.predict(Xnew_test)
score_kn = accuracy_score(ynew_test,y_predict_kn)
print(score_kn)
```

```python
# logistic regression
my_logreg = LogisticRegression(random_state=5)

my_logreg.fit(Xnew_train, ynew_train)
y_predict_lr = my_logreg.predict(Xnew_test)
score_lr = accuracy_score(ynew_test, y_predict_lr)
print(score_lr)
```

```python
# decision tree
my_decisiontree = DecisionTreeClassifier(random_state=5)
my_decisiontree.fit(Xnew_train, ynew_train)
y_predict_dt = my_decisiontree.predict(Xnew_test)
score_dt = accuracy_score(ynew_test, y_predict_dt)
print(score_dt)
```

# H)

```python
# cross validation
from sklearn.model_selection import cross_val_score


# Applying 10-fold cross validation with "logistic regression" classifier:

# In the following line, "my_logreg" is instantiated as an "object" of LogisticRegression "class". 
# my_logreg = LogisticRegression(solver='lbfgs')

# function cross_val_score performs Cross Validation:
accuracy_list_logistic = cross_val_score(my_logreg, Xnew, ynew, cv=10, scoring='accuracy')
accuracy_list_knn = cross_val_score(knn, Xnew, ynew, cv=10, scoring='accuracy')
accuracy_list_decision = cross_val_score(my_decisiontree, Xnew, ynew, cv=10, scoring='accuracy')

# use average of accuracy values as final result
accuracy_cv_lg = accuracy_list_logistic.mean()
accuracy_cv_kn = accuracy_list_knn.mean()
accuracy_cv_dt = accuracy_list_decision.mean()
# printing the acuracy
# Best accuracy is logistic regreesion and worst is Knn
print("Logistic "+ str(accuracy_cv_lg))
print("Knn "+str(accuracy_cv_kn))
print("Decision tree "+str(accuracy_cv_dt))
```
